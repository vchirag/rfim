```python
import numpy as np
DTYPE = np.float32

from numba import njit
import matplotlib.pyplot as plt
import pandas as pd
import os
from tqdm import tqdm
```

```python
"""
T: Temperature
L: Lattice Size
J: Interaction Strength
h: External Field
h_rand: Random local fields
m: magnetization
z: coord. number
"""

@njit
# Function to calculate effective field
def effective_field(i, j, m, h_rand, h, J, z):
    return J*z*m + h_rand[i, j] + h

# Function to calculate the magnetization of the system
@njit
def calculate_magnetization(spins):
    return np.sum(spins)/((np.shape(spins)[0])**2)
```

```python
L = int(100)
J = DTYPE(1)
# External field
h = 0
dh = 0.01

# Random Fields (normally distributed)
std = DTYPE(0.1)
# h_rand = np.random.normal(0, std, size= (L, L))
h_rand = np.random.triangular(-1, 0, +3, size = (L, L)) # triangle dist

# ext h increment steps
h_steps = int(1000)

# Initial config
spins = np.random.choice([-1, 1], size=(L, L))  # Random initial spins

# Mean Field (2D)
z = 4
```

```python
plt.imshow(h_rand)
```

```python
magnetizations = [] # Recording m for every field increment

m = calculate_magnetization(spins)

h = -5.0  # Start with a negative field to align spins

avalanche_sizes = []

for step in tqdm(range(h_steps)):
    unstable = []
    for i in range(L):
        for j in range(L):
            if effective_field(i, j, m, h_rand, h, J, z) * spins[i, j] < 0:
                unstable.append((i, j))

    
    avalanche_size = 0
    while unstable:
        avalanche_size += len(unstable)
        new_unstable = []
        for (i, j) in unstable:    
            spins[i, j] *= -1  # Flip spin
            m = calculate_magnetization(spins)
            
            for (ni, nj) in [(i + 1, j), (i - 1, j), (i, j + 1), (i, j - 1)]:
                ni, nj = ni % L, nj % L
                if effective_field(ni, nj, m, h_rand, h, J, z) * spins[ni, nj] < 0:
                    if (ni, nj) not in unstable and (ni, nj) not in new_unstable:
                        new_unstable.append((ni, nj))
        unstable = new_unstable
    
    if avalanche_size > 0:
        avalanche_sizes.append(avalanche_size)
    h += dh
    magnetizations.append(m)
    
```

```python
avalanche_sizes
```

```python
if avalanche_sizes:
    hist, bins = np.histogram(avalanche_sizes, bins=np.logspace(0, np.log10(max(avalanche_sizes)), 50), density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    plt.loglog(bin_centers, hist, 'o', label="Simulation Data")
    plt.loglog(bin_centers, bin_centers**(-3/2), '--', label="$s^{-3/2}$")
    plt.xlabel("Avalanche Size $s$")
    plt.ylabel("Probability Density $P(s)$")
    plt.title("Avalanche Size Distribution in 2D RFIM")
    plt.legend()
    plt.show()
else:
    print("No avalanches occurred during the simulation.")
```

```python
h = -5
hs = np.arange(h, int(h+h_steps*dh), dh)
plt.plot(hs, magnetizations)
```
